### 3.8 Adding columns for "Total Score" and "Average"

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? ) 

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
print(gender_group)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'gender': ['Male', 'Female', 'Male', 'Female'],
    'average': [75, 80, 85, 90],
    'math score': [78, 82, 88, 94]
})

df.columns = df.columns.str.strip()
gender_group = df.groupby('gender', as_index=True).mean(numeric_only=True)

if 'math score' not in gender_group.columns:
    raise KeyError("Column 'math score' not found in DataFrame. Check column names!")

female_scores = [gender_group.loc['Female', 'average'], gender_group.loc['Female', 'math score']]
male_scores = [gender_group.loc['Male', 'average'], gender_group.loc['Male', 'math score']]

plt.figure(figsize=(10, 8))
X = ['Total Average', 'Math Average']
X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, male_scores, 0.4, label='Male', color='blue')
plt.bar(X_axis + 0.2, female_scores, 0.4, label='Female', color='pink')

plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total Average vs Math Average Marks by Gender", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.

#### 4.4.2 RACE/EHNICITY COLUMN
- How is Group wise distribution ?
- Is Race/Ehnicity has any impact on student's performance ?

#### UNIVARIATE ANALYSIS ( How is Group wise distribution ?)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'race_ethnicity': ['Group A', 'Group B', 'Group A', 'Group C', 'Group B', 'Group C', 'Group A'],
})

df.columns = df.columns.str.strip()

if 'race/ethnicity' in df.columns:
    race_col = 'race/ethnicity'
elif 'race_ethnicity' in df.columns:
    race_col = 'race_ethnicity'
else:
    raise KeyError("Column 'race/ethnicity' not found. Available columns: " + ", ".join(df.columns))

f, ax = plt.subplots(1, 2, figsize=(20, 10))

sns.countplot(x=df[race_col], data=df, palette='bright', ax=ax[0], saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container, color='black', size=20)

ax[1].pie(x=df[race_col].value_counts(), labels=df[race_col].value_counts().index,
          explode=[0.1] + [0] * (len(df[race_col].unique()) - 1),
          autopct='%1.1f%%', shadow=True)

plt.show()
```

#### Insights 
- Most of the student belonging from group C /group D.
- Lowest number of students belong to groupA.
